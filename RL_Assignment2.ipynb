```python
import pandas as pd
import numpy as np
import glob


def load_runs(log_dir):
    files = sorted(glob.glob(f"{log_dir}/seed_*.csv"))
    runs = [pd.read_csv(f) for f in files]
    return runs


def compute_auc(returns):
    return np.mean(returns)  # normalized AUC


def compute_final_performance(returns, k=50):
    return np.mean(returns[-k:])


def compute_metrics(log_dir):
    runs = load_runs(log_dir)

    aucs = []
    finals = []

    for r in runs:
        returns = r["return"].values

        aucs.append(compute_auc(returns))
        finals.append(compute_final_performance(returns))

    metrics = {
        "AUC_mean": np.mean(aucs),
        "AUC_std": np.std(aucs),
        "Final_mean": np.mean(finals),
        "Final_std": np.std(finals),
    }

    return metrics


def compare_all(base_dir="logs/raw"):
    experiments = {
        "trunc_200": f"{base_dir}/trunc_200",
        "trunc_1000": f"{base_dir}/trunc_1000",
        "trunc_2000": f"{base_dir}/trunc_2000",
    }

    results = {}

    for name, path in experiments.items():
        results[name] = compute_metrics(path)

    return results

results = compare_all()

for exp, metrics in results.items():
    print(f"\n=== {exp} ===")
    for k, v in metrics.items():
        print(f"{k}: {v:.2f}")
```
